# Questão 4

```python
# a) Crie um dicionário chamado pessoa com as seguintes chaves e valores: "nome": "Carlos"; "idade": 30; "cidade": "São Paulo".
pessoa = {"nome": "Carlos", "idade": 30, "cidade": "São Paulo"}

# b) Imprima o dicionário pessoa
print("Dicionário:")
print(pessoa)

# c) Acesse e imprima o valor da chave "nome".
print("\nNome:", pessoa["nome"])

# d) Altere e imprima a idade de "Carlos" para 31.
pessoa["idade"] = 31
print("\nDicionário após alterar a idade:")
print(pessoa)

# e) Adicione uma nova chave "profissão" ao dicionário com o valor "engenheiro" e imprima.
pessoa["profissão"] = "engenheiro"
print("\nDicionário após adicionar profissão:")
print(pessoa)

# f) Remova a chave "cidade" do dicionário e imprima o dicionário após a remoção.
pessoa.pop("cidade")
print("\nDicionário após remover cidade:")
print(pessoa)

# g) Utilize o método .keys() para imprimir todas as chaves do dicionário.
print("\nChaves do dicionário:")
print(", ".join(pessoa.keys()))  # Exibe as chaves sem aspas e separadas por vírgula
print(pessoa.keys())

# h) Utilize o método .values() para imprimir todos os valores do dicionário.
print("\nValores do dicionário:")
print(", ".join(map(str, pessoa.values())))  # Exibe os valores sem aspas e separados por vírgula
print(pessoa.values())
```

# Questão 9

```python
# a) Faça um programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o imposto de renda, 8% para o INSS e 5% para o sindicato. O programa deve mostrar: salário bruto, quanto pagou ao INSS, quanto pagou ao sindicato, salário líquido.
ganho_por_hora = float(input("Quanto você ganha por hora? "))
horas_trabalhadas = float(input("Quantas horas você trabalhou no mês? "))
print()

# Cálculos
salario_bruto = ganho_por_hora * horas_trabalhadas
imposto_renda = salario_bruto * 0.11
inss = salario_bruto * 0.08
sindicato = salario_bruto * 0.05
salario_liquido = salario_bruto - imposto_renda - inss - sindicato

# Resultados
print(f"Salário Bruto: R$ {salario_bruto:.2f}")
print()

print(f"Imposto de Renda (11%): R$ {imposto_renda:.2f}")
print()

print(f"INSS (8%): R$ {inss:.2f}")
print()

print(f"Sindicato (5%): R$ {sindicato:.2f}")
print()

print(f"Salário Líquido: R$ {salario_liquido:.2f}")
```

```python
#Importando pacotes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression #Regressão Linear
from sklearn.metrics import r2_score #R²
```

```python
#importando o conjunto de dados
dados = pd.read_csv("C:/Users/20221enpro0308/Downloads/score_updated.csv")

#mostrando dados
dados
```

```python
# 1) Diagrama de Dispersão
plt.scatter(dados['Hours'], dados['Scores'], color = 'lightcoral')
plt.title('Horas de Estudo vs. Pontuação na prova')
plt.xlabel('Horas de Estudo')
plt.ylabel('Pontuação na prova')
```

```python
#Separação de X e Y
x = dados.iloc[:, :1] #variável independente
y = dados.iloc[:, 1:] #variável dependente

# 2) Ajuste do modelo (encontrar beta0 e beta1 que minimizam o erro)
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
print('Coeficiente angular (beta1): ', modelo.coef_)
print('Intercepto (beta0): ', modelo.intercept_)

# modelo: y = 1,863576 + 9,901318x + E
```

```python
# Fazendo Predições para y usando x
y_pred = modelo.predict(x)

#Gráfico
plt.scatter(dados['Hours'], dados['Scores'], color = 'lightcoral')
plt.plot(x, y_pred, color = 'firebrick')
plt.title('Horas de Estudo vs. Pontuação na prova')
plt.xlabel('Horas de Estudo')
plt.ylabel('Pontuação na prova')
```

```python
#Verificação do Coeficiente de Determinação (R²)
r2_score(y, y_pred)
```
